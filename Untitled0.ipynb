```python
import os
import numpy as np
import cv2
import dlib
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array
import matplotlib.pyplot as plt
import json
```

```python
# Load the trained model
model_path = '/content/drive/MyDrive/vgg16/best_model.keras'  # Adjust the path as needed
model = load_model(model_path)

# Path to metadata file
metadata_path = '/content/drive/MyDrive/vgg16/train_sample_videos/metadata.json'

# Load metadata
with open(metadata_path, 'r') as file:
    metadata = json.load(file)
```

```python
# Function to extract frames from a video file and preprocess them
def preprocess_video(video_path, input_shape=(128, 128, 3)):
    cap = cv2.VideoCapture(video_path)
    frames = []
    frame_indices = []

    frameRate = cap.get(cv2.CAP_PROP_FPS)  # Frame rate of the video

    while cap.isOpened():
        frameId = cap.get(1)  # Current frame number
        ret, frame = cap.read()
        if not ret:
            break

        # Process every nth frame based on frame rate
        if frameId % int(frameRate) == 0:
            # Convert the frame to RGB (OpenCV uses BGR by default)
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

            # Resize the frame to the desired input shape
            frame_resized = cv2.resize(frame_rgb, (input_shape[0], input_shape[1]))

            # Convert the frame to an array and normalize it
            frame_array = img_to_array(frame_resized) / 255.0
            frames.append(frame_array)
            frame_indices.append(frameId)

    cap.release()
    return np.array(frames), frame_indices
```

```python
video_path = '/content/drive/MyDrive/vgg16/train_sample_videos/eepezmygaq.mp4'  # Adjust the path as needed

# Preprocess the video
frames, frame_indices = preprocess_video(video_path)
print(f"Extracted {len(frames)} frames from the video.")
'''
ecnihjlfyt
ebchwmwayp
ehtdtkmmli
dtocdfbwca
drcyabprvt
dulanfulol
ccfoszqabv
dgmevclvzy
djvtbgwdcc
bgvhtpzknn real
ecujsjhscd real
dqswpjoepo fake
dzwkmcwkwl fake
bdnaqemxmr real
drcyabprvt real
anpuvshzoo real
'''
```

```python
# Make predictions on the frames
predictions = model.predict(frames)

# Print predictions for debugging
print("Predictions for each frame:")
print(predictions)

# Average the prediction probabilities
average_prediction = np.mean(predictions, axis=0)
predicted_class = np.argmax(average_prediction)

print(f"Average prediction probabilities: {average_prediction}")
print(f"Predicted class based on average probabilities: {'Fake' if predicted_class == 1 else 'Real'}")
```
